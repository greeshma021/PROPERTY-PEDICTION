```python
import pandas as pd
import numpy as  np
```

```python
df=pd.read_csv('HousePrices.csv')
df
```

```python
df.isnull().sum()
```

```python
missing_values=df.columns[df.isnull().any()]

df[missing_values].isnull().sum()
```

```python
df['LotFrontage'].fillna(df['LotFrontage'].mean(),inplace=True)
df['MasVnrArea'].fillna(df['MasVnrArea'].mean(),inplace=True)
```

```python
x=df[['Dwell_Type','LotFrontage', 'LotArea','OverallQual','OverallCond','MasVnrArea',
'BsmtFinSF1','BsmtUnfSF','TotalBsmtSF','1stFlrSF','2ndFlrSF','GrLivArea','GarageArea',
'WoodDeckSF', 'OpenPorchSF']]
y=df['Property_Sale_Price']
```

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
model.fit(x,y)
```

```python
model.score(x,y)
```

```python
house=model.predict([[70,85,12000,6,8,100,520,430,1024,950,850,1800,700,150,42]])
house
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size = 0.25,random_state=0)
```

```python
from sklearn.linear_model import LinearRegression
lin_reg_mod = LinearRegression()
lin_reg_mod.fit(X_train, y_train)
```

```python
pred = lin_reg_mod.predict(X_test)
pred
```

```python
from sklearn import metrics
test_set_r2 = metrics.r2_score(y_test, pred)
print(test_set_r2)
```
